```python
#DECISION TREE - CANCER DATASET
#STEP 1: LOAD THE DATASET
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv("Downloads/cancer_data.csv")
df.head()
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
#STEP 2: SPLIT THE FEATURES AND LABEL
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
#STEP 3: TRAINING AND TEST SPLIT
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)#
```

```python
#STEP 4: DEFINE THE CLASSIFICATION MODEL
model=DecisionTreeClassifier(criterion='entropy',random_state=42)
model.fit(x_train,y_train)
```

```python
#STEP 5: PERFORMANCE EVALUATION
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#STEP 6: CLASSIFY THE NEW SAMPLE DATA
y_pred = model.predict(df.drop('diagnosis', axis=1).iloc[[0]])
print(y_pred)
```

```python
#STEP 7: VISULIZATION
feature_names = df.columns.drop('diagnosis')
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=feature_names, class_names= model.classes_)
plt.title("Decision Tree for cancer disease Classification")
plt.show()
```

```python
#NAIVE BAYES
#STEP 1: LOAD THE DATASET
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
df=pd.read_csv("Downloads/fruits.csv")
df.head()
```

```python
#STEP 2: SPLIT THE FEATURES AND LABEL
x=df[['weight','size','color_score']]
y=df['Fruit_label']
```

```python
#STEP 3: TRAINING AND TEST SPLIT
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#STEP 4: DEFINE THE CLASSIFICATION MODEL
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
#STEP 5: PERFORMANCE EVALUATION
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#STEP 6: CLASSIFY THE NEW SAMPLE DATA
y_pred_sample=model.predict([[150,7.5,0.50]])
print(y_pred_sample)
```

```python
#STEP 7: VISULIZATION
cm=confusion_matrix(y_test,y_pred , labels=model.classes_)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot()
```

```python
#NAVIE BAYES -NB USER DATASET
#STEP 1: LOAD THE DATASET
df=pd.read_csv("Downloads/NBUser data.csv")
df.head()
```
