```python
#BINNING DATA SMOOTHING
#EQUI FREQUENCY BINNING
def equifreq(arr1, m):
    arr1.sort()
    a = len(arr1)
    n = int(a / m)
    for i in range(0, m):
        arr = []
        for j in range(i * n, (i + 1) * n):
            if j >= a:
                break
            arr.append(arr1[j])
        print("Bin", i+1, ":", arr)

data = [5, 10, 11, 15, 13, 35, 50, 55, 72, 92, 204, 215]

m = 3

print("Equal frequency binning")
equifreq(data, m)
```

```python
#EQUI FREQUENCY BINNING BY MEAN
def equifreq(arr1, m):
    arr1.sort()
    a = len(arr1)
    n = int(a / m)
    
    for i in range(0, m):
        arr = []
        for j in range(i * n, (i + 1) * n):
            if j >= a:
                break
            arr.append(arr1[j])
        
        # Calculate mean of the current bin
        mean = sum(arr) / len(arr) 
        
        print("Bin", i+1,":", [mean] * len(arr))

data = [5, 10, 11, 15, 13, 35, 50, 55, 72, 92, 204, 215]
m = 3

print("Equal frequency binning with bin means")
equifreq(data, m)
```

```python
#EQUI FREQUENCY BINNING BY BIN BOUNDARIES
def equifreq(arr1, m):
    arr1.sort()
    a = len(arr1)
    n = int(a / m)

    for i in range(m):
        arr = []
        for j in range(i * n, (i + 1) * n):
            if j >= a:
                break
            arr.append(arr1[j])

    
        min_val = min(arr)
        max_val = max(arr)
        
        # Replace each value with nearest boundary
        replaced = []
        for val in arr:
            if abs(val - min_val) <= abs(val - max_val):
                replaced.append(min_val)
            else:
                replaced.append(max_val)

        print(f"Bin {i+1}: {replaced}")

data = [5, 10, 11, 15, 13, 35, 50, 55, 72, 92, 204, 215]
m = 3

print("Equal frequency binning with values replaced by nearest boundaries")
equifreq(data, m)
```

```python
#EQUI FREQUENCY BINNING BY MEDIAN
def equifreq(arr, m):
    arr.sort()
    a = len(arr)
    n = int(a / m)

    for i in range(m):
        bins = []
        for j in range(i * n, (i + 1) * n):
            if j >= a:
                break
            bins.append(arr[j])
        
        # Calculate median without built-in function
        mid = len(bins) // 2
        if len(bins) % 2 == 0:
            median_val = (bins[mid - 1] + bins[mid]) / 2
        else:
            median_val = bins[mid]
        
        
        print(f"Bin {i + 1}:", [median_val] * len(bins))

data = [5, 10, 11, 15, 13, 35, 50, 55, 72, 92, 204, 215]
m = 3

print("Equal frequency binning with values replaced by median ")
equifreq(data, m)
```
